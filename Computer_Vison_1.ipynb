```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/content/download.jpeg')
```

```python
img
```

```python
img_rgb = cv2.cvtColor (img, cv2.COLOR_BGRA2BGR)
```

```python
img_rgb
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
img_gray
```

```python
kernel_avg = np.ones((5,5), np.float32)/25
kernel_avg
```

```python
img_avg = cv2.filter2D (img_gray, -1, kernel_avg)
```

```python
img_avg
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(img_gray, cmap = 'gray')
plt.title('Gambar asli')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(img_avg, cmap = 'gray')
plt.title('Gambar hasil filter')
plt.axis('off')
plt.show()
```

```python
gaussian_img = cv2.GaussianBlur(img_gray,(5,5),0)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(img_gray, cmap = 'gray')
plt.title('Gambar asli')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(gaussian_img, cmap = 'gray')
plt.title('Gaussian blur')
plt.axis('off')
plt.show()
```

```python
kernel_avg1 = np.ones((5,5), np.float32)/15
kernel_avg1
```

```python
img_avg1 = cv2.filter2D (img_gray, -1, kernel_avg1)
img_avg1
```

```python
sharp_kernel = np.array([[0, -1, 0],[-1, 5, -1],[0, -1, 0]])
```

```python
img_sharp = cv2.filter2D (img_gray, -1, sharp_kernel)
img_sharp
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,3,1)
plt.imshow(img_gray, cmap = 'gray')
plt.title('Gambar asli')
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(img_avg1, cmap = 'gray')
plt.title('average filter')
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(img_sharp, cmap = 'gray')
plt.title('sharpening kernel')
plt.axis('off')
plt.show()
```
